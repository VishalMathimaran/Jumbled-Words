```python
import pandas as pd
```

```python
words = pd.read_json("./words_dictionary.json", orient='index')
```

```python
allwords = pd.DataFrame(words.index)
```

```python
test_str=allwords.iloc[20][0]
all_freq={}
for i in test_str: 
    if i in all_freq: 
        all_freq[i] += 1
    else: 
        all_freq[i] = 1
```

```python
alpha = 'a'
test_list=[]

for i in range(0, 26): 
    test_list.append(alpha) 
    alpha = chr(ord(alpha) + 1) 
main_df=pd.DataFrame()
for word in allwords.values:
    diction={}
    for i in test_list:
        diction[i]=test_str.count(i)
    data = pd.DataFrame.from_dict(diction, orient='index')
    df1_transposed = data.T 
    df1_transposed["name"] = word
    main_df=main_df.append(df1_transposed, ignore_index=True)
```
